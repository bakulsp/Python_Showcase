```python
# Import libraries
import numpy as np
import pandas as pd
```

```python
# Read Source Data
#Read Customer Sheet
bank_customer = pd.read_excel("Credit Banking.xls", sheet_name = "Customer")

#Read Spend Sheet
bank_spend = pd.read_excel("Credit Banking.xls", sheet_name = "Spend")

#Read Repayment Sheet
bank_repayment = pd.read_excel("Credit Banking.xls", sheet_name = "Repayment")
```

```python
# See sample Data| Spend Sheet
bank_spend.head(5)
```

```python
# Derive Mean Age from Customer Sheet
mean_Age = bank_customer['Age'].mean()
mean_Age
```

```python
# Q. 1 a Where ever the age value is less than 18, make it mean of age values.
#Step 1--> Find how many values have Age Value > 18

bank_customer["Age"] = np.where(bank_customer.Age < 18, mean_Age, bank_customer.Age)
```

```python
bank_customer.Age[bank_customer.Age<18].shape
```

```python
bank_customer.head()
```

```python
# Q. 1 b.	Where ever the spend amount is more than limit, 
# replace it with half of the limit for each customer.
```

```python
# Step 1 is to rebuild entire Spend sheet with unique customers and entire spend amount per month for a given year
```

```python
bank_spend['month'] = pd.DatetimeIndex(bank_spend['Date']).month
```

```python
bank_spend['year'] = pd.DatetimeIndex(bank_spend['Date']).year
```

```python
bank_spend.head()
```

```python
bank_spend['month_year'] = pd.to_datetime(bank_spend['Date']).dt.to_period('M')
```

```python
bank_spend.shape
```

```python
bank_merged = pd.merge(bank_customer,bank_spend,on='Customer',how='inner')
bank_merged.shape
```

```python
bank_merged["Amount_Spend"] = np.where(bank_merged.Amount_Spend > bank_merged.Limit, bank_merged.Limit/2, bank_merged.Amount_Spend)
```

```python
bank_merged.head()
```

```python
# Q. c.	Where ever the repayment amount is more than limit, replace it with the limit.
```

```python
# Step 1 is to rebuild entire Repayment sheet with unique customers and entire spend amount per month for a given year
```

```python
bank_repayment['month'] = pd.DatetimeIndex(bank_repayment['Date']).month
```

```python
bank_repayment['year'] = pd.DatetimeIndex(bank_repayment['Date']).year
```

```python
bank_repayment.head()
```

```python
bank_repayment['month_year'] = pd.to_datetime(bank_repayment['Date']).dt.to_period('M')
```

```python
bank_repayment.shape
```

```python
bank_merged1 = pd.merge(bank_customer,bank_repayment,on='Customer',how='inner')
bank_merged1.shape
```

```python
bank_merged1["Amount_Repayment"] = np.where(bank_merged1.Amount_Repayment > bank_merged1.Limit, bank_merged1.Limit, bank_merged1.Amount_Repayment)
```

```python
bank_merged1.head()
```

```python
bank_merged1.reset_index()
```

```python
#Q 2 a	Monthly spend for each customer
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
bank_merged.groupby(['Customer','month_year'])['Amount_Spend'].sum()
```

```python
ans2a=bank_merged.groupby(['Customer','month_year'])['Amount_Spend'].sum()
```

```python
ans2a=ans2a.reset_index()
```

```python
ans2a
```

```python
#Q 2 b	Monthly Repayment for each customer
```

```python
bank_merged1.groupby(['Customer','month_year'])['Amount_Repayment'].sum()
```

```python
ans2b=bank_merged1.groupby(['Customer','month_year'])['Amount_Repayment'].sum()
```

```python
ans2b=ans2b.reset_index()
```

```python
ans2b
```

```python
ans = pd.merge(ans2a,ans2b,on=['Customer','month_year'],how='outer')
ans.shape
```

```python
ans.reset_index()
```

```python
# Find Blank entries for Repayment
ans.Amount_Spend[ans.Amount_Spend.isna()].shape
```

```python
ans["Amount_Spend"] = np.where(ans.Amount_Spend.isna(), 0, ans.Amount_Spend)
```

```python
ans["Amount_Repayment"] = np.where(ans.Amount_Repayment.isna(), 0, ans.Amount_Repayment)
```

```python
ans.reset_index()
```

```python
ans.Amount_Repayment[ans.Amount_Repayment.isna()].shape
```

```python

ans['Monthly_Int']= ((ans.Amount_Spend-ans.Amount_Repayment)*0.029)
```

```python
ans.reset_index()
```

```python
ans["Monthly_Int"] = np.where(ans.Monthly_Int < 0, 0, ans.Monthly_Int)
```

```python
ans.reset_index()
```
