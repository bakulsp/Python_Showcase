```python

import pandas as pd
```

```python
house=pd.read_csv("house.csv")
```

```python
house.head()
```

```python
# Size of the chart
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

# Feed Plot Baseline Data
plt.plot([1,2,3,4],[12.3,20,16,18],label ="Year 2019")
plt.plot([1,2,3,4],[18.3,21,17.2,19],label="Year 2020")

# Add title
plt.title("Line Chart")

# Labels
plt.xlabel("Quarter")
plt.ylabel("Sales Value")

# Add Legend
plt.legend()

plt.subplot(1,2,2)

plt.scatter(house.SqFt,house.Price)
plt.title("Scatter Plot")
plt.show()
```

```python
# Seaborn -- More powerful
import seaborn as sns
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(data=house,x="SqFt",y="Price",hue="Brick")
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(data=house,x="SqFt",y="Price",hue="Neighborhood")
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(data=house,x="SqFt",y="Price",size="Bedrooms",hue="Brick")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.boxplot(data=house,x="Price")

plt.subplot(1,2,2)
sns.boxplot(data=house,y="Price")
plt.show()
```

```python
plt.figure(figsize=(12,4))

sns.boxplot(data=house,x="Neighborhood",y="Price")
plt.show()
```

```python
plt.figure(figsize=(12,4))

sns.barplot(data=house,x="Neighborhood",y="Price")
plt.show()
```

```python
# Histogram
sns.distplot(house.Price,bins=30)
plt.xlabel("Histogram")
plt.xlim(60000,300000)

plt.show()
```

```python
priceyes=house[house.Brick=="Yes"]["Price"]
```

```python
priceno=house[house.Brick=="No"]["Price"]
```

```python
sns.distplot(priceyes,label="Yes")
sns.distplot(priceno,label="No")
plt.legend()

plt.show()
```

```python
house.corr()
```

```python
sns.heatmap(house.corr(),annot=True,cmap="Wistia")
plt.show()
```

```python
sns.heatmap(house.corr(),annot=True,cmap="viridis")
plt.savefig("Data.jpeg")
```

```python
sns.pairplot(house)
plt.savefig("paired.pdf")
```
